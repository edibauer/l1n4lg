We cant get all inverse matrix for all matrix

```python
import numpy as np
```

```python
a = np.array([[3, 5], [2, 8]])
print(a)
```

```python
det_a = np.linalg.det(a)
print(det_a)
```

```python
a_1 = np.array([[8, -5], [-2, 3]])
print(a_1)
```

```python
inv_a = 1/14 * a_1
print(inv_a)
```

```python
f_inv_a = np.linalg.inv(a)
print(f_inv_a)
```

```python
id_a = a @ inv_a
print(id_a)
```

```python
f_id_a = np.dot(a, f_inv_a)
print(f_id_a)
```
