```python
import numpy as np
import matplotlib.pyplot as plt
```

### Suma de matrices

```python
v = np.array([3, 2])
w = np.array([1, 3])
r = v + w
r
```

```python
origin = np.array([0,0])
plt.arrow(origin[0], origin[1], v[0], v[1], head_width=0.2, head_length=0.3, length_includes_head=True, color='r')
plt.arrow(origin[0], origin[1], w[0], w[1], head_width=0.2, head_length=0.3, length_includes_head=True, color='b')

plt.arrow(origin[0], origin[1], r[0], r[1], head_width=0.2, head_length=0.3, length_includes_head=True, color='g') # ans

plt.axis([0, 6, 0 ,6])
plt.grid()

plt.show()
```
