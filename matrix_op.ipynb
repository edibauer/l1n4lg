### Adding

```python
a = np.array([1, 2])
b = np.array([4, 5])

r = a + b
print(r)
```

```python
c = np.array([3, 2])
d = np.array([4, -1])

r = c + d
print(r)
```

```python
e = np.array([[3, 2], [1, 4]])
print(e)
```

```python
d = np.array([[5, 2],[3, 1]])
print(d)
```

```python
ans = e + d

print(ans)
```

```python
f = np.array([[1, 2], [5, 2], [3, 0]])
print(f)
```

```python
g = np.array([[6, 2], [1, 5], [1, 5]])
print(g)
```

```python
ans2 = f + g
print(ans2)
```

### Substracting

```python
h = np.array([7, 8])
print(h)
```

```python
i = np.array([2, 6])
print(i)
```

```python
ans_res = h - i
print(ans_res)
```

### Mult

ROW * COLUMN

Use a dot product to do multiplication

```python
j = np.array([[1, 2], [3, 4], [5, 6]])
print(j)
```

```python
k = np.array([[2, 3, 2], [5, 1, 2]])
print(k)
```

```python
mult_ans = j @ k
print(mult_ans)
```

```python
mult_ans2 = np.dot(j, k)
print(mult_ans2)
```
