### Proc
When we do a scalar product, we "change the size of a vector"

```python
v = np.array([3, 2])
```

```python
print(v)
```

```python
r = 3 * v
```

```python
print(r)
```

```python
from turtle import color


origin = np.array([0,0])

plt.arrow(origin[0], origin[1], v[0], v[1], head_width=0.3, head_length=0.3, color='r')
plt.arrow(origin[0], origin[1], r[0], r[1], head_width=0.3, head_length=0.3, color='g')

plt.axis([0, 12, 0, 12])
plt.grid()
plt.show()

```
