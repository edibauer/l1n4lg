```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v = np.array([2, 5])
print(v)
```

```python
w = np.array([3, 4])
print(w)
```

```python
dot_vw = v[0] * w[0] + v[1] * w[1]
print(dot_vw)
```

```python
acum = 0

for i in range(len(v)):
    acum += v[i] * w[i]
print(acum)
```
